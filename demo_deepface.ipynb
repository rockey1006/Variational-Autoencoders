```
from skimage import data, io

img = io.imread('/content/need/000027.png')
io.imshow(img)

io.show()
```

```
img = io.imread('/content/need/000027/5.png')
io.imshow(img)

io.show()
```

```
from deepface import DeepFace
result  = DeepFace.verify("/content/need/000027.png", "/content/need/000027/5.png",model_name="Facenet",distance_metric='euclidean_l2')

print("Is verified by facenet: ", result["verified"])
print("euclidean_l2 distance: ",result["distance"])
## for Eucledian < 1.1 is same person
```
